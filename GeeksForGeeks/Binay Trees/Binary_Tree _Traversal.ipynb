```python
#TRIAL
#9-Inorder Successor of a node in Binary Tree
#search node.data by inorder traversal then get successor
#Morris Traversal :: We could have global PREV var. and each time current will be stored to PREV 
#if PREV is Key then Current is inorder successor of PREV/KEY

#Node 
class Node:
    def __init__(self,data):
        self.right=None
        self.left=None
        self.data=data

#Morris Traversal Inorder
def MorisTraversal(root,key):
    if root is None:
        return     
    current=root
    
    while current:
        prev=current 
        if prev==key:  #inorder successor wont be immediate left or right node. Specially in right cubtree
            return current
            
        if current.left is None:
            current=current.right
            
        else:
            pre=current.left
             
            while pre.right is not None and pre.right is not current:
                pre=pre.right
             
            if pre.right is None:
                pre.right=current
                current=current.left
                
            else:
                pre.right=None
                current=current.right
                
#main
# Driver code
"""
Constructed binary tree is
            1
          /   \
         2     3
       /   \
      4     5
"""

root = Node(10)
root.left = Node(2)
root.left.right=Node(5)
root.left.left=Node(4)
root.left.right.right=Node(8)
root.right = Node(3)
root.right.left=Node(6)
root.right.right=Node(7)
root.right.right.left=Node(9)

for v in MorisTraversal(root,4):
    print(v, end=' ')
```

```python
#6.find all possible Binary Trees with the given Inorder traversal and print their preorder traversals
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
def preorder(root):
    if root is not None:
        print(root.data, end=" ")
        preorder(root.left)
        preorder(root.right)
        
# Function for constructing all possible trees with
# given inorder traversal stored in an array from
# arr[start] to arr[end]. This function returns a
# vector of trees.
def getTrees(arr,start,end):  #finding permutation of all possible trees inorder
    # List to store all possible trees
    trees=[]
    
    """ if start > end then subtree will be empty so
    returning NULL in the list """
    if start>end:
        trees.append(None)
        return trees
    
    """ Iterating through all values from start to end
        for constructing left and right subtree
        recursively """
    for i in range(start,end+1):    #this is logic for permutation of list/trees
        # Constructing left subtree
        ltrees=getTrees(arr,start,i-1)
        
         # Constructing right subtree
        rtrees=getTrees(arr,i+1,end)
        
        """ Looping through all left and right subtrees
        and connecting to ith root below"""
        for j in ltrees:
            for k in rtrees:
                # Making arr[i]  as root
                curr_root=Node(arr[i])
                
                # Connecting left subtree
                curr_root.left=j
                curr_root.right=k
                
                #feeding root to tree array
                trees.append(curr_root)
    return trees

# Driver program to test above function
inp=[4,5,7]
n=len(inp)

trees=getTrees(inp,0,n-1)
print("Preorder traversals of different possible Binary Trees are")
for root in trees:  #here compare with preorders
    preorder(root)
    print(" ")
```

```python
#10-Find n-th node of inorder traversal
#Moris Traversal with Count
class newNode:
    def __init__(self,data):
        self.right=None
        self.left=None
        self.data=data
        
def morisTraversal(root, n):
    if root is None:
        return
    
    current=root
    cnt=0
    while current:
       
        
        if cnt==n:
            return current.data
        
        if current.left is None:
            cnt+=1
            current=current.right
            
        else:
            pred=current.left
            
            while pred.right is not current and pred.right is not None:
                pred=pred.right
            
            
            if pred.right is None:
                pred.right=current
                current=current.left
            else:
                pred.right is None
                cnt+=1
                current=current.right
                
    if cnt==n:
        return current.data
    else:
        print("tree nodes are less in number")
        return
            
root = newNode(10)
root.left = newNode(20)
root.right = newNode(30)
root.left.left = newNode(40)
root.left.right = newNode(50)
print(morisTraversal(root,3))
```

```python
#12-Level Order Binary Tree Traversal
#14-Level order traversal line by line 
#Method 01 - Recursive
#Node class
class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
#traversal
def printData(root,level):    
    if root is None:
        return
    
    if level==1:
        print(root.data,end=" ")
    else:
        printData(root.left,level-1)
        printData(root.right,level-1)
        
#calculate height
def height(root):
    if root is None:
        return 0
    else:
        lheight=height(root.left)
        rheight=height(root.right)
        
    if lheight>rheight:
        return lheight+1
    else:
        return rheight+1
    
#utility function
def levelOrder(root):
    h=height(root)
    for i in range(1,h+1):
        printData(root,i)
        
#driver method
# Driver program to test above function
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
 
print("Level order traversal of binary tree is -")
levelOrder(root)
```

```python
#method 02-Using Queue
#class
class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
#Traversal Using Q
def levelOrder(root):
    q=[]  #we gonna use q to track nodes by level
    
    if root is None:
        return
    
    q.append(root)  #first element in q should be at level 1
    
    while len(q):
        temp=q[0]  #assigning first element in q to temp
        q.pop(0)
        
        print(temp.data, end=" ")
        
        if temp.left is not None:
            q.append(temp.left)
        
        if temp.right is not None:
            q.append(temp.right)
    
#main Method
# Driver program to test above function
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
 
print("Level order traversal of binary tree is -")
levelOrder(root)
```

```python
#21-Morris traversal for Preorder
#Node
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
#Moris Traversal
def morisPreorder(root):
    if root is None:
        return
    
    current=root  #travenrse 
    
    while current:
        if current.left is None:  #only right nodes
            print(current.data, end=" ")
            current=current.right        
        
        else:
            prev=current.left
            
            while prev.right is not None and prev.right is not current:#moving prev only when finding successor
                prev=prev.right  #preorder successor  
                
            if prev.right is None:
                print(current.data, end=" ")
                prev.right=current
                current=current.left
                
            else:                
                prev.right =None
                current=current.right  #always moving current            
        
#Recursive traversal
def recPreorder(root):
    if root is None :
        return
    
    print(root.data, end=" ")
    recPreorder(root.left)
    recPreorder(root.right)
    
#main 
root = Node(1)
root.left = Node(2)
root.right = Node(3)
  
root.left.left = Node(4)
root.left.right = Node(5)
  
root.right.left= Node(6)
root.right.right = Node(7)
  
root.left.left.left = Node(8)
root.left.left.right = Node(9)
  
root.left.right.left = Node(10)
root.left.right.right = Node(11)
   
morisPreorder(root)
print("\n")
recPreorder(root)
```
