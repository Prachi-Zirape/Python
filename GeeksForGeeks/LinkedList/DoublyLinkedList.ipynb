```python
#quick sort revision :;divide and conquer method
#method chose any element as pivot 
def partition(arr,low,high):
    pivot=arr[high]  #last element as picot
    
    i=low-1
    # Traverse through all elements
    # compare each element with pivot
    for j in range(low,high):
        if arr[j]<=pivot:
             # If element smaller than pivot is found
            # swap it with the greater element pointed by i
            i=i+1
             # If element smaller than pivot is found
            # swap it with the greater element pointed by i
            (arr[i],arr[j])=(arr[j],arr[i])
            
    #pushing pivot element in a way tthat left elelment pivot>left and pivot<right
    (arr[high],arr[i+1])=(arr[i+1],arr[high])
    return i+1

def quick_sort(array, low, high):
    if low < high:
  
        # Find pivot element such that
        # element smaller than pivot are on the left
        # element greater than pivot are on the right
        pi = partition(array, low, high)
  
        # Recursive call on the left of pivot
        quick_sort(array, low, pi - 1)
  
        # Recursive call on the right of pivot
        quick_sort(array, pi + 1, high)
  
  
# Driver code
array = [2,10, 7, 8, 9, 1, 5]
quick_sort(array, 0, len(array) - 1)
  
print(f'Sorted array: {array}')
```

```python
#11-Insert value in sorted way in a sorted doubly linked list
# 3 cases 1-head, 2-end 3-in between
#node
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
#ll
class DoublLinkedlist:
    def __init__(self):
        self.head=None
        
    #push
    def push(self,data): #if node is smallest or single
        new_node=Node(data) #created new node
        
        new_node.next=self.head 
        
        if self.head is not None:
            self.head.prev=new_node
        
        self.head=new_node
        return
        
    #insert after
    def insert_after(self,node,data):
        if self.head is None:  #empty list
            return
        
        new_node=Node(data)
        
        new_node.next=node.next
        node.next=new_node
        new_node.prev=node
        new_node.next.prev=new_node
        return
        
    #insert end
    def insert_End(self,data):
        new_node=Node(data)  #new node
        temp=self.head #traverse to end
        
        while temp.next:
            temp=temp.next
            
        new_node.next=temp.next
        temp.next=new_node
        new_node.prev=temp
        
    #insertion in sorted list
    def sortedInsertion(self,data):        
        #case1- its head
        if data<self.head.data:
            self.push(data)
            return
        
        #case 2 end
        temp=self.head
        while temp:
            temp=temp.next
            
            if (temp.data<data and temp.next is None):
                self.insert_End(data)
                break
                return
            
            elif (temp.next.data>data and temp.next.next is not None):
                self.insert_after(temp,data) 
                break
                return
            
    #print
    def printList(self):
        temp=self.head
        while temp:
            print(temp.data,end=" ")
            last=temp
            temp=temp.next
        print()
        while last:
            print(last.data,end=" ")
            last=last.prev
            
#main
dll=DoublLinkedlist()
dll.push(100)
dll.push(90)
dll.push(80)
dll.push(70)
dll.push(60)
dll.push(50)
dll.push(40)
dll.printList()
print()
dll.insert_End(110)
dll.printList()
print()
dll.insert_after(dll.head.next,55)
dll.printList()
print()
dll.sortedInsertion(39)
dll.printList()
print()
dll.sortedInsertion(53)
dll.printList()
print()
dll.sortedInsertion(110)
dll.printList()
```

```python
#12-Delete a Doubly Linked List node at a given position
#Node
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
#ll
class DoubLinkedList:
    def __init__(self):
        self.head=None
    #push
    def push(self,data):
        new_node=Node(data)
        
        new_node.next=self.head
        if self.head is not None:
            self.head.prev=new_node
            
        self.head=new_node        
    #print
    def printList(self):
        temp=self.head
        while temp:
            print(temp.data, end=" ")
            temp=temp.next
    #delete
    def delete_given_pos(self,k):
        cnt=1
        temp=self.head
        
        if self.head.next is None:
            return 
        
        if self.head is None:
            return
        
        while cnt!=k:
            cnt+=1
            temp=temp.next
        
        if (cnt>1 and temp.next is not None):
            temp.prev.next=temp.next
            temp.next.prev=temp.prev
        elif (cnt==1):
            self.head=temp.next
        elif temp.next is None:
            temp.prev.next=None     
        
#main
dll=DoubLinkedList()
dll.push(10)
dll.push(20)
dll.push(30)
dll.push(40)
dll.push(50)
dll.push(60)
dll.printList()
dll.delete_given_pos(6)
print()
dll.printList()
```
