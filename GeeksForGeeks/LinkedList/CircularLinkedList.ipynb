```python
#2-Traversal/printing of CLL

#create node
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class CircLinkedList:
    #create lst node
    def __init__(self):
        self.last=None

    #add to empty list
    def addToEmpty(self,new_data):
        #create node
        new_node=Node(new_data)
        
        #see if list has node already
        if self.last!=None:
            return self.last
        
        self.last=new_node   #assign to new node
        new_node.next=self.last  #link to self
        
        return self.last
    
    #add at beginniig
    def addToBegin(self,data):       
        #check if list is empty or not
        if self.last==None:
            return addToEmpty(data)
        
         #create new node
        new_node=Node(data)
        
        #pushback first node
        new_node.next=self.last.next
        
        #link new node with previous node
        self.last.next=new_node
        return self.last        
        
    #add at end
    def addToend(self,data):
        #see if list is empty
        if self.last==None:
            return addToEmpty(data)
        #create new Node
        new_node=Node(data)
        
        #pushback last node
        new_node.next=self.last.next  
        self.last.next=new_node
        self.last=new_node  #additional step to define last
        return self.last
    
    #add after
    def addAfter(self,new_data,given_data):
        #see if list is empty
        if self.last==None:
            return addToEmpty(new_data)
        
        #create new node
        new_node=Node(new_data)
        
        first_node=self.last.next
        
        while(first_node):
            if first_node.data==given_data:
                new_node.next=first_node.next  #pushing back node next to given node
                first_node.next=new_node   #adding new node after given node
            
            if first_node==self.last.next:
                break
            
            first_node=first_node.next
        return self.last
    
    #printList
    def printList(self):
        temp=self.last.next
        while(temp):
            print(temp.data,end=" ")
            temp=temp.next
            if temp==self.last.next:
                break                

#main
cllist=CircLinkedList()
cllist.addToEmpty(10)
cllist.addToBegin(20)
cllist.addToBegin(30)
cllist.addToend(50)
cllist.addAfter(100,30)
cllist.printList()
```

```python
#4-Sorted insert for circular linked list
#Node
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
#llist
class CircLinkedList:
    def __init__(self):
        self.head=None
    #print
    def print_list(self):
        temp=self.head
        while(temp):
            print(temp.data)
            temp=temp.next
            if temp==self.head:
                break
    
    #insert
    def insert(self,data):
        #create node
        new_node=Node(data)
        temp=self.head
        
        new_node.next=self.head  #pushing back head
        
        if self.head!=None:
            while(temp.next!=self.head):
                 temp=temp.next
            temp.next=new_node   #inserting at end
        else:
            new_node.next=new_node   #single node
        self.head=new_node      #assign new node as head
            
            
    #key sorted insert
    def sorted_insertion(self,key):
        #create node with key
        key_node=Node(key)
        temp=self.head
        
        #CLl empty
        if self.head is None:            
            key_node.next=key_node
            self.head=key_node
        
        #node is smaller than head
        elif temp.data>=key_node.data:
            # If value is smaller than head's value then we
            # need to change next of last node
            while temp.next!=self.head:
                temp=temp.next
            temp.next=key_node
            key_node.next=self.head
            self.head=key_node
            
        #node is greater than head
        else:
            while(temp.next!=self.head and temp.next.data<key_node.data):
                temp=temp.next
            key_node.next=temp.next
            temp.next=key_node                       
        
#main
cllist=CircLinkedList()
cllist.insert(70)
cllist.insert(60)
cllist.insert(40)
cllist.insert(10)
cllist.sorted_insertion(50)
cllist.sorted_insertion(80)
cllist.sorted_insertion(0)
cllist.print_list()
```
