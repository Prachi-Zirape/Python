```python
#8-Search an element in a Linked List (Recursive)
#create Node
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
#create LL
class LinkedList:
    def __init__(self):
        self.head=None    
        
    #print ll
    def printList(self):
        temp=self.head
        while(temp):
            print(temp.data, end=" ")
            temp=temp.next
            
    #search recursively
    def rec_search(self,node,x):
        if not node:
            return False
        if node.data==x:
            return True
        return self.rec_search(node.next,x)       
    
#main
llist=LinkedList()
first=Node(1)
second=Node(2)
third=Node(3)
forth=Node(4)
fifth=Node(5)

llist.head=first
first.next=second
second.next=third
third.next=forth
forth.next=fifth

llist.printList()

llist.rec_search(first,4)
```
