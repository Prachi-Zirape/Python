```python
#Bubble Sort Algorithm -swapping the adjacent elements if they are in the wrong order
A = [64, 25, 12, 22, 11, 65]

def bubbleSort(arr):
    for i in range(len(A)):
        swapped=False
        for j in range(0,len(A)-i-1):  #cause every swap largest element go at end
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swapped=True
        
        # IF no two elements were swapped
        # by inner loop, then break
        if swapped==False:
            break
            
bubbleSort(A)
print ("Sorted array :")
for i in range(len(A)):
    print ("%d" %A[i],end=" ")
```

```python
#Bubble Sort Algorithm-Recursive
class bubbleSort:
    def __init__(self,array):
        self.array=array
        self.length=len(array)
    
    def __str__(self):
        return " ".join([str(x) for x in self.array])
    
    def bubbleSortRecursive(self,n=None):
        if n is None:
            n=self.length
        count=0
        
        #base case
        if n==1:
            return
        # One pass of bubble sort. After
        # this pass, the largest element
        # is moved (or bubbled) to end.
        for i in range(n-1):
            if self.array[i]>self.array[i+1]:
                self.array[i],self.array[i+1]=self.array[i+1],self.array[i]
                count+=1
        # Check if any recursion happens or not
          # If any recursion is not happen then return
        if count==0:
            return
        # Largest element is fixed,
        #  recur for remaining array
       # print(n)
        self.bubbleSortRecursive(n-1)
        
def main():
    array = [64, 34, 25, 12, 22, 11, 90]
      
    # Creating object for class
    sort = bubbleSort(array)
      
    # Sorting array
    sort.bubbleSortRecursive()
    print("Sorted array :\n", sort)
  
  
if __name__ == "__main__":
    main()          
```

```python
#Quick Sort
def partition(array,low,high):
    # Choose the rightmost element as pivot
    pivot=array[high]   #last elelemnt
    
    # Pointer for greater element
    i=low-1
    
    # Traverse through all elements
    # compare each element with pivot
    for j in range(low,high):
        if array[j]<=pivot:
            # If element smaller than pivot is found
            # swap it with the greater element pointed by i
            i+=1
            
            # Swapping element at i with element at j
            (array[i],array[j])=(array[j],array[i])
            
    # Swap the pivot element with
    # e greater element specified by i
    array[i+1],array[high]=array[high],array[i+1]
   # print(pivot,array,i)
    return i+1

def quick_sort(array,low,high):
    if low<high:
        
        # Find pivot element such that
        # element smaller than pivot are on the left
        # element greater than pivot are on the right
        pi=partition(array,low,high)

        # Recursive call on the left of pivot
        quick_sort(array,low,pi-1)
        
        # Recursive call on the right of pivot
        quick_sort(array,pi+1,high)
        
array = [10, 7, 8, 9, 1, 5]
quick_sort(array, 0, len(array) - 1)
 
print(f'Sorted array: {array}') 
```

```python
#Iterative quick sort
def partition(arr,lo,hi):
    #mark pivot
    pivot=arr[hi]
    
    #mark index to swap pivot
    i=lo-1
    
    #start swapping elelemnts that are smaller than pivot
    for j in range(lo,hi):
        if arr[j]<=pivot:
            #increase i by 1 cause there is no increment to i
            i+=1
            
            arr[j],arr[i]=arr[i],arr[j]           
            
    #time to swap pivot with greater element
    arr[hi],arr[i+1]=arr[i+1],arr[hi]
    #pivot,arr[i+1]=arr[i+1],pivot
    return i+1

def quick_sort_iter(arr,lo,hi):
    # Create an auxiliary stack
    size=hi-lo+1
    stack=[0]*size
    
    #initialize top of the stack
    top=-1
    
    # push initial values of l and h to stack
    top=top+1
    stack[top]=lo
    top=top+1
    stack[top]=hi
    
    # Keep popping from stack while is not empty
    while top>=0:
        #pop h and lo
        hi=stack[top]
        top=top-1
        lo=stack[top]
        top=top-1
        
        # Set pivot element at its correct position in
        # sorted array
        p=partition(arr,lo,hi)
        
        # If there are elements on left side of pivot,
        # then push left side to stack
        if p-1>lo:
            top=top+1
            stack[top]=lo
            top=top+1
            stack[top]=p-1
            
        # If there are elements on right side of pivot,
        # then push right side to stack
        if p+1<hi:
            top=top+1
            stack[top]=p+1
            top=top+1
            stack[top]=hi
arr = [4, 3, 5, 2, 1, 3, 2, 3]
n = len(arr)
quick_sort_iter(arr, 0, n-1)
print ("Sorted array is:")
for i in range(n):
    print (arr[i],end=" ")   
```

```python
#Date-9/22/22
#heap Sort-repeatedly convert heap to max heap. and replace removed max heap element with root elelment with last root element

# To heapify subtree rooted at index idx.
# N is size of heap
def heapify(arr,N,idx):
    #print(idx)
    largest=idx ## Initialize largest as root
    l=2*idx+1   # left = 2*i + 1
    r=2*idx+2   #right = 2*i + 2
    
    #both ifs will be checked as its not else if
    
    # See if left child of root exists and is
    # greater than root
    if l<N and arr[largest]<arr[l]:
        largest=l
    
    # See if right child of root exists and is
    # greater than root
    if r<N and arr[largest]<arr[r]:
        largest=r
        
    # Change root, if needed()
    if largest!=idx:
        arr[idx],arr[largest]=arr[largest],arr[idx]
        
        # Heapify the root
        heapify(arr,N,largest)
        
def heap_sort(arr):
    N=len(arr)
    
    # Build a maxheap.
    for i in range(N//2-1,-1,-1):
        heapify(arr,N,i)
        #print(arr,i)
    
    # One by one extract elements
    for i in range(N-1,0,-1):
        arr[i], arr[0] = arr[0], arr[i]  # swap
        #print(arr[i],arr[0])
        heapify(arr, i, 0)  #recursively pushin largest elment to end and reducing idx by 1
    
arr = [12, 11, 13, 5, 6, 7]
 
# Function call
heap_sort(arr)
N = len(arr)
 
print("Sorted array is")
for i in range(N):
    print("%d" % arr[i], end=" ")  
    
```

```python
#counting sort

def countSort(arr):
         # Find the maximum element in the inputArray
        maxElement=max(arr)
        
        countArrayLen=maxElement+1
        
        # Initialize the countArray with (max+1) zeros
        countArray=[0]*countArrayLen
        
        # Step 1 -> Traverse the inputArray and increase 
    # the corresponding count for every element by 1
        for el in arr:
            countArray[el]+=1
        #print("cntArr=" ,countArray)
        
        # Step 2 -> For each element in the countArray, 
    # sum up its value with the value of the previous 
    # element, and then store that value 
    # as the value of the current element
        for i in range(1,countArrayLen):
            countArray[i]+=countArray[i-1]
        #print("cntArr=" ,countArray)
        
        # Step 3 -> Calculate element position
    # based on the countArray values
        outputArray=[0]*len(arr)
        i=len(arr)-1
        
        print("cntArr=" ,countArray)
        while i>=0:
            currentEl=arr[i]
           # print("currentEl",currentEl)
            countArray[currentEl]-=1
            newPosition=countArray[currentEl]            
            outputArray[newPosition]=currentEl
            i-=1
        return outputArray
    
inputArray = [2,2,0,6,1,9,9,7]
print("Input array = ", inputArray)

sortedArray = countSort(inputArray)
print("Counting sort result = ", sortedArray)      
```

```python
#Radix Sort
# The idea of Radix Sort is to do digit by digit sort starting from least 
#significant digit to most significant digit. Radix sort uses counting sort as a subroutine to sort.

def countingSort(arr,exp1):
    n=len(arr)
    
    # The output array elements that will have sorted arr
    output=[0]*n
    
    # initialize count array as 
    count=[0]*10
    
    # Store count of elements in count[]
    for i in range(0,n):
        idx=arr[i]//exp1  #exp is 10's index
        count[idx%10]+=1
        
    # Change count[i] so that count[i] now contains actual
    # position of this digit in output array
    for i in range(1,10):
        count[i]+=count[i-1]
        
    ## Build the output array
    i=n-1
    while i>=0:
        idx=arr[i]//exp1
        output[count[idx%10]-1]=arr[i]
        count[idx%10]-=1
        i-=1
        
    # Copying the output array to arr[],
    # so that arr now contains sorted numbers
    for i in range(0,len(arr)):
        arr[i]=output[i]
        
#Radix sort
def radixSort(arr):
    # Find the maximum number to know number of digits
    max1=max(arr)
    # Do counting sort for every digit. Note that instead
    # of passing digit number, exp is passed. exp is 10^i
    # where i is current digit number
    exp=1
    while max1/exp>=1:
        countingSort(arr, exp)
        exp *= 10

arr = [170, 45, 75, 90, 802, 24, 2, 66]
 
# Function Call
radixSort(arr)
 
for i in range(len(arr)):
    print(arr[i],end=" ")
```

```python
#bucket sort for decimals having integer part
# Bucket sort for numbers 
# having integer part
def bucketSort(arr,numOfBuckets):
    max_el=max(arr)
    min_el=min(arr)
    
    # range(for buckets)
    rng=(max_el-min_el)/numOfBuckets
    
    temp=[]
    
    # create empty buckets
    for i in range(numOfBuckets):
        temp.append([])
        
    # scatter the array elements
    # into the correct bucket
    for i in range(len(arr)):
        diff=(arr[i]-min_el)/rng-int((arr[i]-min_el)/rng)
        
    # append the boundary elements to the lower array
        if (diff==0 and arr[i]!=min_el):
            temp[int((arr[i]-min_el)/rng)-1].append(arr[i])
        else:
            temp[int((arr[i]-min_el)/rng)].append(arr[i])
    
    # Sort each bucket individually
    for i in range(len(temp)):
        if len(temp[i]) != 0:
            temp[i].sort()
            
    #concatenate all the buckets
    k=0
    for lst in temp:
        if lst:
            for i in lst:
                arr[k]=i
                k+=1
                
arr = [9.8, 0.6, 10.1, 1.9, 3.07, 3.04, 5.0, 8.0, 4.8, 7.68]
noOfBuckets = 5
bucketSort(arr, noOfBuckets)
print("Sorted array: ", arr)
```

```python
#tim Sort
# Python3 program to perform basic timSort
MIN_MERGE = 32 
 
def calcMinRun(n):
    """Returns the minimum length of a
    run from 23 - 64 so that
    the len(array)/minrun is less than or
    equal to a power of 2.
 
    e.g. 1=>1, ..., 63=>63, 64=>32, 65=>33,
    ..., 127=>64, 128=>32, ...
    """
    r = 0
    while n >= MIN_MERGE:
        r |= n & 1
        n >>= 1
    return n + r 
 
# This function sorts array from left index to
# to right index which is of size atmost RUN
def insertionSort(arr, left, right):
    for i in range(left + 1, right + 1):
        j = i
        while j > left and arr[j] < arr[j - 1]:
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            j -= 1
  
# Merge function merges the sorted runs
def merge(arr, l, m, r):
 
    # original array is broken in two parts
    # left and right array
    len1, len2 = m - l + 1, r - m
    left, right = [], []
    for i in range(0, len1):
        left.append(arr[l + i])
    for i in range(0, len2):
        right.append(arr[m + 1 + i])
 
    i, j, k = 0, 0, l
 
    # after comparing, we merge those two array
    # in larger sub array
    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
 
        else:
            arr[k] = right[j]
            j += 1
 
        k += 1
 
    # Copy remaining elements of left, if any
    while i < len1:
        arr[k] = left[i]
        k += 1
        i += 1
 
    # Copy remaining element of right, if any
    while j < len2:
        arr[k] = right[j]
        k += 1
        j += 1 
 
# Iterative Timsort function to sort the
# array[0...n-1] (similar to merge sort)
def timSort(arr):
    n = len(arr)
    minRun = calcMinRun(n)
 
    # Sort individual subarrays of size RUN
    for start in range(0, n, minRun):
        end = min(start + minRun - 1, n - 1)
        insertionSort(arr, start, end)
 
    # Start merging from size RUN (or 32). It will merge
    # to form size 64, then 128, 256 and so on ....
    size = minRun
    while size < n:
 
        # Pick starting point of left sub array. We
        # are going to merge arr[left..left+size-1]
        # and arr[left+size, left+2*size-1]
        # After every merge, we increase left by 2*size
        for left in range(0, n, 2 * size):
 
            # Find ending point of left sub array
            # mid+1 is starting point of right sub array
            mid = min(n - 1, left + size - 1)
            right = min((left + 2 * size - 1), (n - 1))
 
            # Merge sub array arr[left.....mid] &
            # arr[mid+1....right]
            if mid < right:
                merge(arr, left, mid, right)
 
        size = 2 * size
 
 
# Driver program to test above function
if __name__ == "__main__":
 
    arr = [-2, 7, 15, -14, 0, 15, 0,
           7, -7, -4, -13, 5, 8, -14, 12]
 
    print("Given Array is")
    print(arr)
 
    # Function Call
    timSort(arr)
 
    print("After Sorting Array is")
    print(arr)    
```
